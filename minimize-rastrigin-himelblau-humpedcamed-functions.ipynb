```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
from mpl_toolkits.mplot3d import Axes3D

def Rastrigin(x, y):
    return 20 + x * x + y * y - 10 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))

def Himmelblau(x, y):
    return (((x**2+y-11)**2) + (((x+y**2-7)**2)))

def ThreehumpedCamel(x1, x2):
    return (2*(x1**2)-1.05*(x1**4)+((x1**6)/6)+(x1*x2)+(x2**2))

def create_element(minimum, maximum):
    return [minimum + random() * (maximum - minimum), minimum + random() * (maximum - minimum)]

def create_start_population(minimum, maximum, size):
    return [create_element(minimum, maximum) for _ in range(size)]

def calculate_fitness(population, fitness):
    return [fitness(population[i][0], population[i][1]) for i in range(len(population))]

def change_population(old_population, length):
    return old_population[:length]

def sort_tuple(p_tuple):
    p_tuple = sorted(p_tuple, key=lambda item: item[1])
    return p_tuple

def create_child(population, p_lambda):
    result = []
    for item in population:
        for _ in range(p_lambda):
            result.append([item[0] + (np.random.randn() * 2), item[1] + (np.random.randn() * 2)])
    return result

def graph(population, left, right, fitness, step,colorr):
    x, y = np.mgrid[left:right:step, left:right:step]
    z = fitness(x, y)
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.view_init(60, 0)
    ax.plot_wireframe(x, y, z,color=colorr)
    px = [population[i][0] for i in range(len(population))]
    py = [population[i][1] for i in range(len(population))]
    pz = [fitness(px[i], py[i]) for i in range(len(population))]
    ax.scatter(px, py, pz, s=100, c='r', marker='o')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

def optimization(fitness, min_value, max_value, p_mu, p_lambda, step, colorr):
    population = create_start_population(min_value, max_value, p_mu)
    for i in range(200):
        fitness_value = calculate_fitness(population, fitness)
        min_fitness = min(fitness_value)
        avg_fitness = sum(fitness_value) / len(fitness_value)
        population += create_child(population, p_lambda)
        fitness_value = calculate_fitness(population, fitness)
        population_tuple = [(population[i], fitness_value[i]) for i in range(len(population))]
        population_tuple = sort_tuple(population_tuple)
        for idx, item in enumerate(population_tuple):
            element, fitness_v = item
            population[idx] = element
            fitness_value[idx] = fitness_v
        population = change_population(population, p_mu)
        if (i % 10 == 0):
            print('Iteration',i)
            print('min =',min_fitness)
            print('avg =',avg_fitness)
            graph(population, min_value, max_value, fitness, step,colorr)

p_mu = 10
p_lambda = 15
print('Rastrigin:')
optimization(Rastrigin, -5.12, 5.12, p_mu, p_lambda, 0.1,"c")
print('\n\nHimmelblau:')
optimization(Himmelblau, -5, 5, p_mu, p_lambda, 0.5,"y")
print('\n\nThreehumpedCamel:')
optimization(ThreehumpedCamel, -5, 5, p_mu, p_lambda, 0.5,"g")
```
