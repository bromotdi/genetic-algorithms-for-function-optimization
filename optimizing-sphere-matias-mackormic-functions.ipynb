```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np

mpl.rcParams['legend.fontsize'] = 10

def functionMackormic(x, y):
    return np.sin(x + y) + (x - y)**2 - 1.5 * x + 2.5 * y + 1

def method(total_iter, n, ls, rs, phenotype):
    number_of_population = 0
    previous_avg = 0
    previous_max = 0
    current_avg = 0
    current_max = 0
    begin_avg = 0
    begin_max = 0
    for i in range(total_iter):
        number_of_population += 1
        print("\nPopulation {0}".format(number_of_population))
        [fitness, current_avg, current_max] = current_population(n, phenotype)
        if i != 0:
            print("\nDifference in population fitness: {0}".format(current_avg - previous_avg))
            print("Difference in max fitness: {0}".format(current_max - previous_max))
            if current_avg - previous_avg == 0 and current_max - previous_max == 0:
                graficMackormic(number_of_population, phenotype, fitness)
                print("\nOptimum for this experiment has been found")
                break
            previous_avg = current_avg
            previous_max = current_max
        else:
            previous_avg = current_avg
            previous_max = current_max
            begin_avg = current_avg
            begin_max = current_max
        graficMackormic(number_of_population, phenotype, fitness)
        if i < total_iter-1:
            phenotype = future_population(n, previous_avg, previous_max, ls, rs, phenotype)
    print("\nDifference in population fitness between first and final population: {0}".format(round(current_avg-begin_avg, 2)))
    print("Difference in max fitness between first and final population: {0}".format(current_max-begin_max))

def current_population(n, phenotype):
    fitness = []
    for i in phenotype:
        fitness.append(functionMackormic(i[0], i[1]))
    print("| {0:^2} | {1:^18} | {2:^18} | {3:^16} |".format("№", "Phenotype(x)", "Phenotype(y)", "Fitness function"))
    for i in range(n):
        print("| {0:^2} | {1:^18} | {2:^18} | {3:^16} |".format(i+1, round(phenotype[i][0], 2), round(phenotype[i][1], 2), round(fitness[i], 2)))
    print("\nPopulation fitness:")
    print(sum(fitness)/n)
    print("\nMax fitness:")
    print(max(fitness))
    return [fitness, sum(fitness)/n, max(fitness)]

def future_population(n, previous_avg, previous_max, ls, rs, phenotype):
    previous_avg_in_process = previous_avg
    previous_max_in_process = previous_max
    current_avg_in_process = 0
    current_max_in_process = 0
    for i in range(len(phenotype)):
        best_organizm = phenotype[0]
        for j in range(1, len(phenotype)):
            if functionMackormic(phenotype[j][0], phenotype[j][1]) > functionMackormic(best_organizm[0], best_organizm[1]):
                best_organizm = phenotype[j]               
                best_organizm_index = j
        print("\nElement in population: {0}".format(i+1))
        phenotype = mutualism(n, ls, rs, i, best_organizm, phenotype)
        print("\nAfter mutualism:")
        [fitness1, current_avg_in_process, current_max_in_process] = current_population(n, phenotype)
        print("\nDifference in population fitness: {0}".format(current_avg_in_process - previous_avg_in_process))
        print("Difference in max fitness: {0}".format(current_max_in_process - previous_max_in_process))
        previous_avg_in_process = current_avg_in_process
        previous_max_in_process = current_max_in_process
        phenotype = comensalism(n, ls, rs, i, best_organizm, phenotype)
        print("\nAfter comensalism:")
        [fitness1, current_avg_in_process, current_max_in_process] = current_population(n, phenotype)
        print("\nDifference in population fitness: {0}".format(current_avg_in_process - previous_avg_in_process))
        print("Difference in max fitness: {0}".format(current_max_in_process - previous_max_in_process))
        previous_avg_in_process = current_avg_in_process
        previous_max_in_process = current_max_in_process
        phenotype = parasitism(n, i, best_organizm, phenotype)
        print("\nAfter parasitism:")
        [fitness1, current_avg_in_process, current_max_in_process] = current_population(n, phenotype)
        print("\nDifference in population fitness: {0}".format(current_avg_in_process - previous_avg_in_process))
        print("Difference in max fitness: {0}".format(current_max_in_process - previous_max_in_process))
        previous_avg_in_process = current_avg_in_process
        previous_max_in_process = current_max_in_process
    return phenotype 

def mutualism(n, ls, rs, first_organizm_index, best_organizm, phenotype):
    second_organizm_index = np.random.randint(0, n)
    while second_organizm_index == first_organizm_index:
        second_organizm_index = np.random.randint(0, n)
    mutual_vector = []
    for i in range(len(phenotype[first_organizm_index])):
        mutual_vector.append((phenotype[first_organizm_index][i] + phenotype[second_organizm_index][i]) / 2)
    BF1 = np.random.randint(1, 3)
    BF2 = np.random.randint(1, 3)
    new_first_organizm = []
    new_second_organizm = []
    for i in range(len(phenotype[first_organizm_index])):
        posible_var_first = phenotype[first_organizm_index][i] + round(np.random.uniform(0, 2), 2) * (best_organizm[i] - mutual_vector[i] * BF1)
        if posible_var_first > rs:
            new_first_organizm.append(63)
        elif posible_var_first < ls:
            new_first_organizm.append(-63)
        else:
            new_first_organizm.append(posible_var_first)
        posible_var_second = phenotype[second_organizm_index][i] + round(np.random.uniform(0, 2), 2) * (best_organizm[i] - mutual_vector[i] * BF2)
        if posible_var_second > rs:
            new_second_organizm.append(63)
        elif posible_var_second < ls:
            new_second_organizm.append(-63)
        else:
            new_second_organizm.append(posible_var_second)
    first_organizm_fitness = functionMackormic(new_first_organizm[0], new_first_organizm[1])
    second_organizm_fitness = functionMackormic(new_second_organizm[0], new_second_organizm[1])
    if first_organizm_fitness > functionMackormic(phenotype[first_organizm_index][0], phenotype[first_organizm_index][1]):
        phenotype[first_organizm_index] = new_first_organizm
    if second_organizm_fitness > functionMackormic(phenotype[second_organizm_index][0], phenotype[second_organizm_index][1]):
        phenotype[second_organizm_index] = new_second_organizm
    return phenotype

def comensalism(n, ls, rs, first_organizm_index, best_organizm, phenotype):
    second_organizm_index = np.random.randint(0, n)
    while second_organizm_index == first_organizm_index:
        second_organizm_index = np.random.randint(0, n)
    new_first_organizm = []
    for i in range(len(phenotype[first_organizm_index])):
        posible_var_first = phenotype[first_organizm_index][i] + round(np.random.uniform(-1, 2), 2) * (best_organizm[i] - phenotype[second_organizm_index][i])
        if posible_var_first > rs:
            new_first_organizm.append(63)
        elif posible_var_first < ls:
            new_first_organizm.append(-63)
        else:
            new_first_organizm.append(posible_var_first)
    first_organizm_fitness = functionMackormic(new_first_organizm[0], new_first_organizm[1])
    if first_organizm_fitness > functionMackormic(phenotype[first_organizm_index][0], phenotype[first_organizm_index][1]):
        phenotype[first_organizm_index] = new_first_organizm
    return phenotype

def parasitism(n, first_organizm_index, best_organizm, phenotype):
    second_organizm_index = np.random.randint(0, n)
    while second_organizm_index == first_organizm_index:
        second_organizm_index = np.random.randint(0, n)
    parasite_vector = []
    for i in range(len(phenotype[first_organizm_index])):
        parasite_vector.append(phenotype[first_organizm_index][i])
    if functionMackormic(parasite_vector[0], parasite_vector[1]) > functionMackormic(phenotype[second_organizm_index][0], phenotype[second_organizm_index][1]):
        phenotype[second_organizm_index] = parasite_vector
    return phenotype

def graficMackormic(number_of_population, phenotype, fitness):
    x, y = np.mgrid[-63:63:1, -63:63:1]
    z = functionMackormic(x, y)
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    ax.set_title("Population {0}".format(number_of_population))
    ax.plot_wireframe(x, y, z, color="yellow", label="Mackormic function")
    for i in range(len(phenotype)):
        ax.plot(phenotype[i][0], phenotype[i][1], fitness[i], "ro")
    ax.legend()
    plt.show()

total_iter = 10
N = 20
left_side = -63
right_side = 63
in_phenotype = []
for i in range(N):
    in_phenotype.append([np.random.randint(left_side, right_side+1), np.random.randint(left_side, right_side+1)])

method(total_iter, N, left_side, right_side, in_phenotype)
```
